```python
import numpy as np
import torch
import torch.nn as nn
from torch.autograd import Variable

# 生成示例数据
def generate_data(n_samples, n_steps):
    X = np.random.randn(n_samples, n_steps, 1)
    y = np.sin(np.arange(n_steps) / 10.0) + np.random.randn(n_steps) * 0.1
    return X, y

# 创建LSTM模型
class LSTMModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(LSTMModel, self).__init__()
        self.hidden_dim = hidden_dim
        self.lstm = nn.LSTM(input_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        h0 = torch.zeros(1, x.size(0), self.hidden_dim).to(x.device)
        c0 = torch.zeros(1, x.size(0), self.hidden_dim).to(x.device)
        out, _ = self.lstm(x, (h0, c0))
        out = self.fc(out[:, -1, :])
        return out

# 准备数据
n_samples = 3000
n_steps = 100
X, y = generate_data(n_samples, n_steps)

```

```python
X[0].shape
```

```python

X_train, y_train = X[:2000], y[:2000]  # 使用前2000条数据作为训练集
X_test, y_test = X[2000:], y[2000:]    # 使用后1000条数据作为测试集

X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32)

# 构建和训练模型
input_dim = 1
hidden_dim = 50
output_dim = 1
model = LSTMModel(input_dim, hidden_dim, output_dim)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

for epoch in range(10):
    model.train()
    optimizer.zero_grad()
    output = model(X_train_tensor[:, :n_steps])
    loss = criterion(output, y_train_tensor.unsqueeze(1))
    loss.backward()
    optimizer.step()
    print('Epoch [{}/{}], Loss: {:.4f}'.format(epoch+1, 10, loss.item()))


# 使用模型进行预测
future_steps = 1000
future_input = X_test_tensor[0].unsqueeze(0)  # 使用测试集的第一个时间步作为初始输入
predictions = []
with torch.no_grad():
    model.eval()
    for i in range(future_steps):
        pred = model(future_input)
        predictions.append(pred.item())
        future_input = torch.cat((future_input[:, 1:, :], pred.unsqueeze(1)), dim=1)

# 打印预测结果
print(predictions)
```
